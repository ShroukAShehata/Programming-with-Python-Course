## 9) Handling Errors

```python
try:
    print(var)
except:
    print("error occured")
finally:
    print("Good Luck")
```

```python
var = 6

try:
    print(var)
except:
    print("error occured")
finally:
    print("Good Luck")
```

```python
try:
    print(z+c)
except(NameError):
    print("Variable is not defined")
except:
    print("Error Occured")
finally:
    print("Good Luck")
```

```python
z = 5
c = "Hello"

try:
    print(z+c)
except(NameError):
    print("Variable is not defined")
except:
    print("Error Occured")
finally:
    print("Good Luck")
```

## 10) OOP

```python
class Member:
    def __init__(self):
        print("New Member has been added")
        
member1 = Member()
member2 = Member()
member3 = Member()
```

```python
class Member:
    def __init__(self):
        self.name = "John"
        
member1 = Member()
member2 = Member()

print(member1.name)
print(member2.name)
```

```python
class Member:
    def __init__(self , first_name):
        self.name = first_name
        
member1 = Member("Mark")
member2 = Member("Sally")

print(member1.name)
print(member2.name)
```
